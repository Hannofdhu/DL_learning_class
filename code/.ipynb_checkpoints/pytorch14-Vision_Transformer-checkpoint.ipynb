只考虑了Tansformer的Encoder部分｜这里是分类任务（CV三大任务——检测、识别、分割，此处只讨论识别(recognition)任务）

不同角度理解VIT：

- DNN角度：1.Image2patch. 2.patch2embedding. 

- CNN角度：1.2D convolution over image. 2.flatten the output feature map.

两个角度得到的embedding是一样的，后面有验证。

VIT的系列操作：

- class token embedding

- position embedding

- Transformer Encoder

- classification head

图片的大小可以变化，但块的大小不变。

开头有一个随机初始化的embedding。

# VIT

## 1.Image2patch. 2.patch2embedding.角度

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

def image2emb_naive(image,patch_size,weight):
    #image_size:bs*channel*h*w
    #此函数用于把图像沿像素，以kernel为依据进行切分
    #bs*patch个数*(patch_size*patch_size*input_channel)=patch_depth
    patch = F.unfold(image,\
    kernel_size=patch_size,stride=patch_size).transpose(-1,-2)
    patch_embedding = patch@weight
    return patch_embedding
    
#test code for image2emb
bs,input_channel,height,width = 1,3,8,8
patch_size = 4
#embedding的大小
model_dim = 8
image = torch.randn(bs,input_channel,height,width)
patch_depth = patch_size*patch_size*input_channel
#weight(patch--embedding)映射参数矩阵
weight = torch.randn(patch_depth,model_dim)#patch_depth是卷积核的面积乘以输入通道数，model_dim就是通道数

patch_embedding_naive = image2emb_naive(image,\
                  patch_size,weight)
#bs*patch个数*model_dim
print(patch_embedding_naive.shape)
print(patch_embedding_naive)
```

## 1.2D convolution over image. 2.flatten the output feature map.角度

```python
def image2emb_conv(image,kernel,stride):
    #bs*output_channel*output_height*output_width
    conv_output = F.conv2d(image,kernel,stride = stride)
    bs,oc,oh,ow = conv_output.shape
    print('oh',oh)
    print('ow',ow)
    patch_embedding_conv = conv_output.reshape((bs,oc,oh*ow)).transpose(-1,-2)
    return patch_embedding_conv
#关键是定义好kernel
kernel =weight.transpose(0,1).reshape(-1,input_channel,patch_size,patch_size) #output_channel*input_channel*kernel_height*kernel_width
print('Kernel shape:',kernel.shape)
patch_embedding_conv = image2emb_conv(image,kernel,patch_size)

print(patch_embedding_conv.shape)
print(patch_embedding_conv)
```

## 开头增加可学习embedding

```python
#CLS token embedding
max_num_token = 16
num_classes = 10
label = torch.randint(10,(bs,))

cls_token_embedding = torch.randn(bs,\
                    1,model_dim,requires_grad=True)
token_embedding = torch.cat([cls_token_embedding,patch_embedding_conv],dim=1)
print(token_embedding.shape)
#加上可训练位置embedding,这里是初始化得到的
position_embedding_table = torch.randn(max_num_token,model_dim,requires_grad=True)
seq_len = token_embedding.shape[1]
position_embedding =torch.tile(position_embedding_table[:seq_len],[token_embedding.shape[0],1,1])
token_embedding +=position_embedding
print(token_embedding.shape)
```

## Embedding送入Transformer Encoder

```python
encoder_layer = nn.TransformerEncoderLayer(d_model=model_dim,nhead=8)
transformer_encoder = nn.TransformerEncoder(encoder_layer,num_layers=6)
encoder_output = transformer_encoder(token_embedding)
print(encoder_output.shape)

#取出第一个位置上的状态[cls]
cls_token_output = encoder_output[:,0,:]
linear_layer = nn.Linear(model_dim,num_classes)
logits = linear_layer(cls_token_output)
print('logits:',logits.shape)
#交叉熵已经把softmax融入其中
loss_fn = nn.CrossEntropyLoss()
loss = loss_fn(logits,label)
print(loss)
```
